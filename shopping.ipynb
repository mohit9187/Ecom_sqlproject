orders_on_different_price_interval

```python
sql ="""
SELECT paymenttype,
SUM(CASE WHEN totalprice >= 0 and totalprice < 10
THEN 1 ELSE 0 END) as cnt_0_10,
SUM(CASE WHEN totalprice >= 10 and totalprice < 100
THEN 1 ELSE 0 END) as cnt_10_100,
SUM(CASE WHEN totalprice >= 100 and totalprice < 1000
THEN 1 ELSE 0 END) as cnt_100_1000,SUM(CASE WHEN totalprice > 1000 
THEN 1 ELSE 0 END) as cnt_1000,sum(totalprice) as revenue
FROM orders
GROUP BY paymenttype
ORDER BY 1
  """
orders_on_different_price_interval= pd.io.sql.read_sql(sql,conn)
orders_on_different_price_interval
```
